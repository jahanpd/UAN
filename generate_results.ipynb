```python
from scipy.stats import t

# n2 is test set size and n1 is training set size
def corrected_pvalue(array1, array2, n2, n1):
    assert len(array1) == len(array2)
    y = len(array1)
    diff = array1 - array2
    mean = diff.mean()
    s2 = diff.var()
    s2mod = (1/y + n2/n1) * s2
    ts = mean / np.sqrt(s2mod)
    print(ts)
    return (1 - t.cdf(ts, y-1))
```
